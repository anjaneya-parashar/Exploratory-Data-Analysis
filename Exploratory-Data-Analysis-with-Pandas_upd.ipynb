```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
women=data['sex']=='Female'
wm=data[women]
wm.head()
WomenAge=wm['age']
WomenAgeMean=WomenAge.mean()
print(WomenAgeMean)
```

**3. What is the proportion of German citizens (*native-country* feature)?**

```python
german_count=data['native-country']=='Germany'
ger=data[german_count]
german_lenght=len(ger)
data_lenght=len(data)
propotion=german_lenght/data_lenght
print(propotion)
```

**4-5. What are mean value and standard deviation of the age of those who recieve more than 50K per year (*salary* feature) and those who receive less than 50K per year? **

```python
c=(data['salary']=='>50K')
Richavg=data[c] 
AgeOfRich=Richavg['age']

meanAgeOfRich=AgeOfRich.mean()
StdAgeOfRich=AgeOfRich.std()
print("mean age of rich is", meanAgeOfRich)
print("std of rich is", StdAgeOfRich)
print("************************")
a=(data['salary']=='<=50K')
PoorAvg=data[a] 
AgeOfPoor=PoorAvg['age']
meanAgeOfPoor=AgeOfPoor.mean()
StdAgeOfPoor=AgeOfPoor.std()
print("mean age of poor is", meanAgeOfPoor)
print("std of poor is", StdAgeOfPoor)
```

**6. Is it true that people who receive more than 50k have at least high school education? (*education - Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
# list of education levels considered high school or above
high_edu = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']

# filter high-earning people
rich = data[data['salary'] == '>50K']

# check if all rich people have high school or above edu
result = rich['education'].isin(high_edu).all()

print("Do all >50K earners have at least high-school education?", result)

# also count those who don't meet the criteria
not_high_edu = rich[~rich['education'].isin(high_edu)]
print("Number of rich people without high school education:", len(not_high_edu))
not_high_edu.head()
```

**7. Display statistics of age for each race (*race* feature) and each gender. Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
# 1. Display age statistics for each race and gender
print(data.groupby(['race', 'sex'])['age'].describe())

# 2. Find maximum age of men from Amer-Indian-Eskimo race
amer_indian_men = data[(data['race'] == 'Amer-Indian-Eskimo') & (data['sex'] == 'Male')]

max_age = amer_indian_men['age'].max()
print("\nMaximum age of Amer-Indian-Eskimo men:", max_age)
```

**8. Among whom the proportion of those who earn a lot(>50K) is more: among married or single men (*marital-status* feature)? Consider married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
# Filter only men
men = data[data['sex'] == 'Male']

# Define married men (marital-status starts with "Married")
married_men = men[men['marital-status'].str.startswith('Married')]

# Define single (bachelor) men
single_men = men[~men['marital-status'].str.startswith('Married')]

# Proportion of >50K earners among married men
married_rich_ratio = (married_men['salary'] == '>50K').mean()

# Proportion of >50K earners among single men
single_rich_ratio = (single_men['salary'] == '>50K').mean()

print("Proportion of rich ( >50K ) among married men:", married_rich_ratio)
print("Proportion of rich ( >50K ) among single men:", single_rich_ratio)
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours and what is the percentage of those who earn a lot among them?**

```python
# Maximum hours worked per week
max_hours = data['hours-per-week'].max()
print("Maximum hours per week:", max_hours)

# People who work that many hours
max_workers = data[data['hours-per-week'] == max_hours]
print("Number of people working that many hours:", len(max_workers))

# Percentage of them who earn >50K
rich_max_workers = (max_workers['salary'] == '>50K').mean() * 100
print("Percentage of people earning >50K among them:", rich_max_workers, "%")
```
